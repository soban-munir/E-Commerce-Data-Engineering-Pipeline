```python
catalog_name = "ecommerce_catalog"
schema_name = "ecom_schema"

spark.sql(f"USE CATALOG ecommerce_catalog")
spark.sql(f"USE SCHEMA ecom_schema")

print(f"Catalog: ecommerce_catalog and Schema: ecom_schema")
```

```python
def register_bronze_table(folder_name, table_name):
    storage_account = "customerprofile" 
    container = "ecommerce-data"
    path = f"abfss://{container}@{storage_account}.dfs.core.windows.net/bronze/{folder_name}/"
    
    print(f"Processing data from folder: {folder_name}...")
    
    df = spark.read.parquet(path)
    df.write.format("delta").mode("overwrite").saveAsTable(f"{catalog_name}.{schema_name}.{table_name}")
    
    print(f"Table '{table_name}' created in Unity Catalog!")
```

```python
register_bronze_table("sales", "raw_sales")
register_bronze_table("customers", "raw_customers")
register_bronze_table("api_users", "raw_api_users")
```

```python
display(spark.table(f"{catalog_name}.{schema_name}.raw_sales").limit(5))
```

```python
df_sales = spark.table(f"{catalog_name}.{schema_name}.raw_sales")
df_customers = spark.table(f"{catalog_name}.{schema_name}.raw_customers")
df_api_users = spark.table(f"{catalog_name}.{schema_name}.raw_api_users")
print("Dataframes created")
```
